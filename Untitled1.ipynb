```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Teacher_Quality"] = df["Teacher_Quality"].fillna(df["Teacher_Quality"].mode()[0])
```

```python
df["Parental_Education_Level"] = df["Parental_Education_Level"].fillna(df["Parental_Education_Level"].mode()[0])
```

```python
df["Distance_from_Home"] = df["Distance_from_Home"].fillna(df["Distance_from_Home"].mode()[0])
```

```python
df.select_dtypes(include=["object"])
```

```python
quality_map = {"Low": 0, "Medium": 1, "High": 2}
```

```python
ord_cols = ["Parental_Involvement", "Access_to_Resources", "Motivation_Level", "Family_Income", "Teacher_Quality", ]
```

```python
for col in ord_cols:
    df[col] = df[col].map(quality_map)
```

```python
df["Parental_Education_Level"].value_counts()
```

```python
edu_map = {"High School": 0, "College": 1,"Postgraduate": 2}
```

```python
df["Parental_Education_Level"] = df["Parental_Education_Level"].map(edu_map)
```

```python
df.select_dtypes(include=["object"])
```

```python
df["Distance_from_Home"].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
feature_orders =[
    ["Public", "Private"],
    ["Negative", "Neutral", "Positive"],
    ["Far", "Moderate", "Near"]
]
```

```python
oe = OrdinalEncoder(categories = feature_orders) 
```

```python
df[["School_Type", "Peer_Influence", "Distance_from_Home"]] = oe.fit_transform(df[["School_Type", "Peer_Influence", "Distance_from_Home"]])
```

```python
df.head()
```

```python
oe2 = OrdinalEncoder()
```

```python
df[["Extracurricular_Activities", "Internet_Access", "Learning_Disabilities", "Gender"]] = oe2.fit_transform(df[["Extracurricular_Activities", "Internet_Access", "Learning_Disabilities", "Gender"]])
```

```python
df.head()
```

```python
X = df.drop(["Exam_Score","Gender"], axis=1)
```

```python
X.columns
```

```python
y = df["Exam_Score"]
```

```python
y
```

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mae
```

```python
r2
```

```python
importance_scaled = lr.coef_
importance_df_scaled = pd.DataFrame({'Feature': X.columns, 'Impact': importance_scaled})
importance_df_scaled = importance_df_scaled.sort_values(by='Impact', ascending=False)

# 5. Plot the new chart
plt.figure(figsize=(10, 8))
sns.barplot(x='Impact', y='Feature', data=importance_df_scaled, palette='viridis')
plt.title('Standardized Importance: The "Fair" Comparison')
plt.show()
```
